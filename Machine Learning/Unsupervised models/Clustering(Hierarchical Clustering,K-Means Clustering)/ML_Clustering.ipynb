## Outlier Treatement

```python
num=[]
cat=[]
for i in Tel_data.columns:
    if Tel_data[i].dtype!='object':
        num.append(i)
    else:
        cat.append(i)
```

## OUTLIER TREATMENT

```python
for i in num:
    ax=sns.boxplot(data=Tel_data,x=i)
    ax.set(title=f"{i} Distribution",xlabel=f"{i} Charge",ylabel="No of customers")
    plt.show()
```

```python
# now we put all the outliers at upper range and lower range 
# when in the outliers there is not distribution like 200,500,900 like this the distribution is more in this outliers so no 
# need to make that outliers in the upper and lower range and if the distribution is less between the ouliers then put it into the
# upper and lower range.
```

```python
Tel_num_data=Tel_data[num]
```

```python
Tel_num_data
    
```

```python
def remove_outliers(col):
    sorted(col)
    q1,q3=np.percentile(col,[25,75])
    IQR=q3-q1
    lr=q1-(1.5 * IQR)
    ur=q3+(1.5 * IQR)
    return lr,ur
```

```python
for col in Tel_num_data:
    lr,ur=remove_outliers(Tel_num_data[col])
    Tel_num_data[col]=np.where(Tel_num_data[col]>ur,ur,Tel_num_data[col])
    Tel_num_data[col]=np.where(Tel_num_data[col]<lr,lr,Tel_num_data[col])
    
```

```python
for i in num:
    ax=sns.boxplot(data=Tel_num_data,x=i)
    ax.set(title=f"{i} Distribution",xlabel=f"{i} Charge",ylabel="No of customers")
    plt.show()
```

## Second method to remove the outliers

```python
for col in num:
    q1=Tel_data[col].quantile(0.25)
    q3=Tel_data[col].quantile(0.75)
    IQR=q3-q1
    lr=q1-(1.5 * IQR)
    ur=q3+(1.5 * IQR)
    Tel_data[col]=np.where(Tel_data[col]>ur,ur,Tel_data[col])
    Tel_data[col]=np.where(Tel_data[col]<lr,lr,Tel_data[col])
```

```python
for i in num:
    ax=sns.boxplot(data=Tel_data,x=i)
    ax.set(title=f"{i} Distribution",xlabel=f"{i} Charge",ylabel="No of customers")
    plt.show()
```

# Clustering

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
wardlink = linkage(Tel_data, method= 'ward')
```

```python
plt.figure(figsize=(10,6))
dendrogram(wardlink,truncate_mode='lastp',p=10);
```

```python
plt.figure(figsize=(10,6))
dendrogram(wardlink,truncate_mode='lastp',p=100);
# how we decide how many clusters we have to take e.g here we take 4 clusters bcz here see if we see last 10 clusters at that
# time also it is showing 4 colors and if we see last 100 clusters at that time also it is showing 4 colors so it means in this
# data 4 clusters are there.
```

```python
plt.figure(figsize=(10,6))
dendrogram(wardlink,truncate_mode='lastp',p=4);
# 1st cluster = 53+55 = 108
# 2nd cluster = 96+41+42 = 179
# 3rd cluster = 79+95 = 174
# 4th cluster = 35+85 = 206
```

```python
# Now we see the clusters wardlink is used to calculate the distance between the points based on that the cluster will created
# Now we want to see that which row belong to which cluster for that we use fcluster.
from scipy.cluster.hierarchy import fcluster
```

```python
# 4 means 4 clusters, criterion means we want maxclust
# fcluster basically use to put every point in the cluster
# here we take the 4 clusters bcz the dendogram shows us the 4 different color if we go for 3 so it will show same 
# color so dendogram show us that use 4 clusters
Hie_cluster = fcluster(wardlink,4,criterion='maxclust')
```

```python
Hie_cluster
# here in array it is showin that 1st row belongs to 2 nd cluster, 2nd row belongs to 3 rd cluster, 3rd row belongs to 1st 
# cluster and so on.
```

```python
len(Hie_cluster)
```
