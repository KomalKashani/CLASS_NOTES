```python
Tel_num_data=df[num]
```

```python
def remove_outliers(col):
    sorted(col)
    q1,q3=np.percentile(col,[25,75])
    IQR=q3-q1
    lr=q1-(1.5 * IQR)
    ur=q3+(1.5 * IQR)
    return lr,ur
```

```python
for col in Tel_num_data:
    lr,ur=remove_outliers(Tel_num_data[col])
    Tel_num_data[col]=np.where(Tel_num_data[col]>ur,ur,Tel_num_data[col])
    Tel_num_data[col]=np.where(Tel_num_data[col]<lr,lr,Tel_num_data[col])
    
```

# Logistic Regression

### Why we use sigmoid function in linear regression for classification and tell that logistic regression.
In logistic regression, we are interested in predicting a binary outcome (e.g., yes/no, true/false, 0/1). The challenge with using linear regression for this type of problem is that linear regression can predict values outside the range of 0 and 1, which aren't suitable for probabilities.

Why Use a Sigmoid Function?
The sigmoid (or logistic) function is used to convert the output of a linear regression into a probability. The output of linear regression is a continuous value, ranging from -âˆž to +âˆž, but probabilities must lie between 0 and 1. The sigmoid function is a non-linear transformation that "squeezes" the output of the linear equation into the range (0, 1), which is perfect for representing probabilities.

The Sigmoid Function Formula:
    Ïƒ(z)= 1/1+e^-z
    
Where z is the output from the linear regression, typically:

z=Î²0 +Î²1x1+Î²2x2+â‹¯+Î²nxn

Here, x_1, x_2, ..., x_n are the features of the input data, and \beta_0, \beta_1, ..., \beta_n are the coefficients (parameters) learned from the training data.

Why it Gives Probability:
The sigmoid function takes any real number and transforms it into a value between 0 and 1.
When z is a large positive number (i.e., the model is highly confident the output is positive), the sigmoid function will output a value close to 1.
When z is a large negative number (i.e., the model is highly confident the output is negative), the sigmoid function will output a value close to 0.
For values of z near 0, the sigmoid function outputs a value near 0.5, indicating uncertainty.
This property allows logistic regression to predict probabilities between 0 and 1, which are interpreted as the likelihood of a particular class (usually class 1).

### Example:
Suppose we want to predict whether an email is spam (1) or not spam (0) based on features like the number of suspicious words, email length, etc. Letâ€™s assume our linear model produces the following value for an email:

### z=2.5

Now, using the sigmoid function:

P(spam)=1/1+e^-2.5 â‰ˆ 0.92

The model predicts a probability of 92% that the email is spam.

If another email produces:

### z=-1.5

Then:
    
P(spam)=1/1+e^-(-1.5) â‰ˆ 0.18

This email has only an 18% chance of being spam, so it is likely to be classified as not spam.

### Summary:
Linear regression alone isn't suitable for binary classification because it can predict values outside [0, 1].
The sigmoid function transforms the linear model's output into the range [0, 1], making it interpretable as a probability.
Logistic regression uses this probability to classify an input as belonging to class 1 (e.g., spam) if the probability is greater than a threshold (usually 0.5), or class 0 (e.g., not spam) otherwise.

### In the sigmoid function, the symbol ð‘’ refers to Euler's number, which is a mathematical constant approximately equal to:

ð‘’â‰ˆ2.718

# AUC - ROC Curve

### roc curve for train data

```python
probs=LR_model.predict_proba(X_train)
# keep probabilities for the positive outcome only
probs = probs[:,1]
# calculate AUC
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(Y_train, probs)
print('AUC: % .2f' % auc)
#calculate roc curve
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(Y_train, probs)
plt.plot([0,1],[0,1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr,tpr,marker='.')
# show the plot
plt.show()
```

### roc curve for test data

```python
probs=LR_model.predict_proba(X_test)
# keep probabilities for the positive outcome only
probs = probs[:,1]
# calculate AUC
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(Y_test, probs)
print('AUC: % .2f' % auc)
#calculate roc curve
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(Y_test, probs)
plt.plot([0,1],[0,1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr,tpr,marker='.')
# show the plot
plt.show()
```

#### The difference between AUC and Roc is not much different and the AUC ranges from 0 to 1.
The AUC value ranges from 0 to 1, where:
AUC = 1: Perfect classifier.
AUC = 0.5: No better than random chance.
AUC < 0.5: Worse than random guessing (the model is making incorrect predictions).

AUC Values and Their Meaning:

AUC = 1.0: Perfect model. It perfectly distinguishes between positive and negative classes at all thresholds.

AUC = 0.9 to 1.0: Excellent model.

AUC = 0.8 to 0.9: Good model.

AUC = 0.7 to 0.8: Fair or moderate model.

AUC = 0.6 to 0.7: Poor model but slightly better than random guessing.

AUC = 0.5: No discrimination (random guessing).

AUC < 0.5: The model is worse than random guessing, which means it is incorrectly ranking negative instances higher than positive ones.

# MODEL REGULARIZATION
Model Regularization means to make a model better.
It is applicable to all the models of classification.

```python
LR_cv = LogisticRegression(solver = 'liblinear')
```

```python
param_grid={
    'penalty':['l1','l2','elasticnet',None],
    'C':[0.01,0.1,1,10,100],
    'class_weight':[None,'balanced']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator=LR_cv,param_grid=param_grid,
                          scoring='f1', cv=5, verbose=1,n_jobs=-1)
```

```python
grid_search.fit(X_train,Y_train)
```

```python
print("Best Parameters found: ",grid_search.best_params_)
print("Best F1 Score: ", grid_search.best_score_)
```

```python
# Now we define the best model the best model is best estimator
best_model = grid_search.best_estimator_
y_pred_train = best_model.predict(X_train)
```

```python
# Confusion matrix
print(metrics.confusion_matrix(Y_train,y_pred_train))
sns.heatmap((metrics.confusion_matrix(Y_train,y_pred_train)),annot=True,fmt='.5g',cmap='Blues');
plt.xlabel('Predicted');
plt.ylabel('Actuals',rotation=0);
plt.title('CONFUSION MATRIX - Cut off (0.5)')
```
