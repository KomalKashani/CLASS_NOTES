```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(20)
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('2017-07-18',periods=10),columns=['a','b','c','d'])
```

```python
df
```

```python
df['new']=df['c']+df['d']
```

```python
df
```

```python
df['new-column-d']=df['d'].apply(lambda x:x**2)
```

```python
df
```

```python
df['new-column-d']=df['d'].apply(lambda x:np.log(x))
```

```python
df
```

```python
 def test(x):
        if x<0:
            return x
        else:
            return np.log(x)
        
```

```python
df['new-column-d-log']=df['d'].apply(test)
```

```python
df
```

```python
 def test1(x):
        if x<0:
            return 1
        else:
            return 0
        
```

```python
df['new-column-d-1']=df['d'].apply(test1)
```

```python
df
```

```python
df.sort_values(by = 'b')
```

```python
df['new_date']=df.index
```
